### Использование свойств даты и времени 

Я хочу работать с датами в столбце datetime как объекты даты и времени вместо простого текста

```python
air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])
```

```python
air_quality["datetime"]
```

Первоначально значения в datetimeявляются символьными строками и не предоставляют никаких операций даты и времени (например, извлечение года, дня недели и т. Д.). Применяя to_datetimeфункцию, pandas интерпретирует строки и преобразует их в объекты datetime (т.е. ). В пандах мы называем эти объекты даты и времени аналогично стандартной библиотеке a .datetime64[ns, UTC]datetime.datetimepandas.Timestamp
